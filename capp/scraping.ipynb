```python
import time
from selenium import webdriver
```

```python
browser = webdriver.Chrome("C:\Program Files (x86)\chromedriver")
```

```python
browser.get("https://www.tripadvisor.in/ShowUserReviews-g304554-d7621418-r558834642-JW_Marriott_Mumbai_Sahar-Mumbai_Maharashtra.html#CHECK_RATES_CONT")
```

```python
rat =[]
ratings =[]
rev = []
reviews = []
date = []
dat =[]
while(len(rev)!=250):
    reviews = [q.text for q in browser.find_elements_by_xpath("//div[@class='entry']")]
    date = browser.find_elements_by_xpath("//div[@class='rating reviewItemInline']")
    ratings = browser.find_elements_by_xpath("//div[@class='rating reviewItemInline']/span[1]")
    c=0
    for i in range(0,len(reviews)):
        if(reviews[i][0:4]!='Dear'):
            rev.append(reviews[i])  
            dat.append(date[c].get_attribute('innerHTML')[93:108].replace("\">R",''))
            rat.append(ratings[c].get_attribute('class')[-2])
            c = c + 1
    try:        
        [r.click() for r in browser.find_elements_by_xpath("//a[@class='nav next taLnk ']")]
        time.sleep(4)
    except:
        None      
    reviews = []
    date =[]
    ratings = []
```

```python
import pandas as pd
import numpy as np
```

```python
l = list(zip(rev,dat,rat))
```

```python
df1 = pd.DataFrame(np.array(l),columns= ('Reviews','Date_of_Review','Ratings'))
df1.Date_of_Review = df1.Date_of_Review.map(lambda x:str(x).replace('R',''))
df1.Date_of_Review = df1.Date_of_Review.map(lambda x:str(x).replace('>',''))
df1.Date_of_Review = df1.Date_of_Review.map(lambda x:str(x).replace('\"',''))
df1.Date_of_Review = df1.Date_of_Review.map(lambda x:str(x).strip())
```
